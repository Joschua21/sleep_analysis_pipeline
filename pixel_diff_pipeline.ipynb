```python
def crop_video_edges(input_video_path, output_video_path=None, 
                    left_cm=0, right_cm=0, top_cm=0, bottom_cm=0,
                    pixels_per_cm=None, reference_width_cm=None, reference_height_cm=None,
                    codec='libx264', quality='high'):
    """
    Crop edges from a video based on centimeter measurements.
    
    Args:
        input_video_path: Path to the input video file
        output_video_path: Path for the output video (optional, auto-generated if None)
        left_cm: Centimeters to crop from left edge
        right_cm: Centimeters to crop from right edge  
        top_cm: Centimeters to crop from top edge
        bottom_cm: Centimeters to crop from bottom edge
        pixels_per_cm: Direct conversion factor (pixels per cm)
        reference_width_cm: Known width of video in cm (for auto-calculation)
        reference_height_cm: Known height of video in cm (for auto-calculation)
        codec: Video codec for output ('libx264', 'libx265', 'mpeg4')
        quality: Output quality ('high', 'medium', 'low')
        
    Returns:
        Path to the cropped video file
    """
    from moviepy.editor import VideoFileClip
    import os
    
    if not os.path.exists(input_video_path):
        raise FileNotFoundError(f"Input video not found: {input_video_path}")
    
    print(f"Loading video: {input_video_path}")
    
    # Load the video
    clip = VideoFileClip(input_video_path)
    original_width, original_height = clip.size
    
    print(f"Original video dimensions: {original_width} x {original_height} pixels")
    print(f"Original video duration: {clip.duration:.2f} seconds")
    
    # Calculate pixels per cm
    if pixels_per_cm is not None:
        # Direct specification
        px_per_cm = pixels_per_cm
        print(f"Using direct conversion: {px_per_cm} pixels per cm")
        
    elif reference_width_cm is not None and reference_height_cm is not None:
        # Calculate from reference dimensions
        px_per_cm_width = original_width / reference_width_cm
        px_per_cm_height = original_height / reference_height_cm
        
        # Use average or warn if they're very different
        px_per_cm = (px_per_cm_width + px_per_cm_height) / 2
        
        if abs(px_per_cm_width - px_per_cm_height) > 5:
            print(f"Warning: Width and height scaling differ significantly:")
            print(f"  Width: {px_per_cm_width:.1f} px/cm")
            print(f"  Height: {px_per_cm_height:.1f} px/cm")
            print(f"  Using average: {px_per_cm:.1f} px/cm")
        else:
            print(f"Calculated conversion: {px_per_cm:.1f} pixels per cm")
            
    elif reference_width_cm is not None:
        # Calculate from width only
        px_per_cm = original_width / reference_width_cm
        print(f"Calculated from width: {px_per_cm:.1f} pixels per cm")
        
    elif reference_height_cm is not None:
        # Calculate from height only  
        px_per_cm = original_height / reference_height_cm
        print(f"Calculated from height: {px_per_cm:.1f} pixels per cm")
        
    else:
        raise ValueError("Must specify either pixels_per_cm OR reference dimensions (width/height in cm)")
    
    # Convert cm measurements to pixels
    left_px = int(left_cm * px_per_cm)
    right_px = int(right_cm * px_per_cm)
    top_px = int(top_cm * px_per_cm)
    bottom_px = int(bottom_cm * px_per_cm)
    
    print(f"\nCropping amounts:")
    print(f"  Left: {left_cm}cm = {left_px}px")
    print(f"  Right: {right_cm}cm = {right_px}px") 
    print(f"  Top: {top_cm}cm = {top_px}px")
    print(f"  Bottom: {bottom_cm}cm = {bottom_px}px")
    
    # Calculate final dimensions
    final_width = original_width - left_px - right_px
    final_height = original_height - top_px - bottom_px
    
    if final_width <= 0 or final_height <= 0:
        raise ValueError(f"Cropping too much! Final dimensions would be {final_width} x {final_height}")
    
    print(f"Final dimensions: {final_width} x {final_height} pixels")
    
    # Apply the crop
    cropped_clip = clip.crop(x1=left_px, y1=top_px, 
                           x2=original_width - right_px, y2=original_height - bottom_px)
    
    # Generate output filename if not provided
    if output_video_path is None:
        base_dir = os.path.dirname(input_video_path)
        base_name = os.path.splitext(os.path.basename(input_video_path))[0]
        crop_suffix = f"_cropped_L{left_cm}_R{right_cm}_T{top_cm}_B{bottom_cm}"
        output_video_path = os.path.join(base_dir, f"{base_name}{crop_suffix}.mp4")
    
    # Set quality parameters with better codec support
    quality_settings = {
        'high': {'bitrate': '8000k', 'preset': 'medium'},
        'medium': {'bitrate': '4000k', 'preset': 'fast'}, 
        'low': {'bitrate': '2000k', 'preset': 'fast'}
    }
    
    settings = quality_settings.get(quality, quality_settings['high'])
    
    print(f"\nSaving cropped video to: {output_video_path}")
    print(f"Codec: {codec}")
    print(f"Quality: {quality} (bitrate: {settings['bitrate']}, preset: {settings['preset']})")
    
    try:
        # Try with the specified codec first
        cropped_clip.write_videofile(
            output_video_path,
            codec=codec,
            bitrate=settings['bitrate'],
            preset=settings['preset'],
            verbose=False,
            logger=None
        )
    except Exception as e:
        print(f"Failed with codec '{codec}': {e}")
        print("Trying fallback codecs...")
        
        # Try fallback codecs
        fallback_codecs = ['libx264', 'mpeg4', 'libx265']
        if codec in fallback_codecs:
            fallback_codecs.remove(codec)
        
        success = False
        for fallback_codec in fallback_codecs:
            try:
                print(f"Trying codec: {fallback_codec}")
                cropped_clip.write_videofile(
                    output_video_path,
                    codec=fallback_codec,
                    bitrate=settings['bitrate'],
                    verbose=False,
                    logger=None
                )
                print(f"Success with codec: {fallback_codec}")
                success = True
                break
            except Exception as fallback_error:
                print(f"Failed with {fallback_codec}: {fallback_error}")
                continue
        
        if not success:
            raise RuntimeError("All codec attempts failed. Please check your FFMPEG installation.")
    
    # Clean up
    clip.close()
    cropped_clip.close()
    
    print(f"Successfully created cropped video: {output_video_path}")
    return output_video_path
```

```python
def crop_video_edges_robust(input_video_path, output_video_path=None, 
                           left_cm=0, right_cm=0, top_cm=0, bottom_cm=0,
                           pixels_per_cm=None, reference_width_cm=None, reference_height_cm=None,
                           quality='high'):
    """
    Crop edges from a video with robust codec handling and validation.
    """
    from moviepy.editor import VideoFileClip
    import os
    
    if not os.path.exists(input_video_path):
        raise FileNotFoundError(f"Input video not found: {input_video_path}")
    
    print(f"Loading video: {input_video_path}")
    
    # Load the video
    clip = VideoFileClip(input_video_path)
    original_width, original_height = clip.size
    
    print(f"Original video dimensions: {original_width} x {original_height} pixels")
    print(f"Original video duration: {clip.duration:.2f} seconds")
    print(f"Original video FPS: {clip.fps}")
    
    # Calculate pixels per cm (same as before)
    if pixels_per_cm is not None:
        px_per_cm = pixels_per_cm
        print(f"Using direct conversion: {px_per_cm} pixels per cm")
    elif reference_width_cm is not None:
        px_per_cm = original_width / reference_width_cm
        print(f"Calculated from width: {px_per_cm:.1f} pixels per cm")
    elif reference_height_cm is not None:
        px_per_cm = original_height / reference_height_cm
        print(f"Calculated from height: {px_per_cm:.1f} pixels per cm")
    else:
        raise ValueError("Must specify either pixels_per_cm OR reference dimensions")
    
    # Convert cm to pixels
    left_px = int(left_cm * px_per_cm)
    right_px = int(right_cm * px_per_cm)
    top_px = int(top_cm * px_per_cm)
    bottom_px = int(bottom_cm * px_per_cm)
    
    print(f"\nCropping amounts:")
    print(f"  Left: {left_cm}cm = {left_px}px")
    print(f"  Right: {right_cm}cm = {right_px}px") 
    print(f"  Top: {top_cm}cm = {top_px}px")
    print(f"  Bottom: {bottom_cm}cm = {bottom_px}px")
    
    # Calculate final dimensions
    final_width = original_width - left_px - right_px
    final_height = original_height - top_px - bottom_px
    
    # Ensure dimensions are even (required by some codecs)
    if final_width % 2 != 0:
        final_width -= 1
        right_px += 1
    if final_height % 2 != 0:
        final_height -= 1
        bottom_px += 1
    
    print(f"Final dimensions (adjusted for codec): {final_width} x {final_height} pixels")
    
    if final_width <= 0 or final_height <= 0:
        raise ValueError(f"Cropping too much! Final dimensions: {final_width} x {final_height}")
    
    # Apply the crop
    cropped_clip = clip.crop(x1=left_px, y1=top_px, 
                           x2=original_width - right_px, y2=original_height - bottom_px)
    
    # Generate output filename
    if output_video_path is None:
        base_dir = os.path.dirname(input_video_path)
        base_name = os.path.splitext(os.path.basename(input_video_path))[0]
        crop_suffix = f"_cropped_L{left_cm}_R{right_cm}_T{top_cm}_B{bottom_cm}"
        output_video_path = os.path.join(base_dir, f"{base_name}{crop_suffix}.mp4")
    
    print(f"\nAttempting to save cropped video to: {output_video_path}")
    
    # Try multiple codec configurations in order of preference
    codec_configs = [
        # Configuration 1: H.264 with high compatibility
        {
            'codec': 'libx264',
            'bitrate': '5000k',
            'audio_codec': 'aac',
            'preset': 'medium',
            'ffmpeg_params': ['-pix_fmt', 'yuv420p']  # Ensures compatibility
        },
        # Configuration 2: H.264 with lower bitrate
        {
            'codec': 'libx264',
            'bitrate': '3000k',
            'audio_codec': 'aac',
            'preset': 'fast',
            'ffmpeg_params': ['-pix_fmt', 'yuv420p']
        },
        # Configuration 3: MPEG-4
        {
            'codec': 'mpeg4',
            'bitrate': '4000k',
            'audio_codec': 'aac'
        },
        # Configuration 4: Simple settings
        {
            'codec': 'libx264',
            'bitrate': '2000k'
        }
    ]
    
    success = False
    for i, config in enumerate(codec_configs):
        try:
            print(f"Trying configuration {i+1}: {config['codec']} codec...")
            
            cropped_clip.write_videofile(
                output_video_path,
                verbose=False,
                logger=None,
                **config
            )
            
            print(f"✓ Success with configuration {i+1}!")
            success = True
            break
            
        except Exception as e:
            print(f"✗ Configuration {i+1} failed: {str(e)[:100]}...")
            if os.path.exists(output_video_path):
                try:
                    os.remove(output_video_path)  # Clean up partial file
                except:
                    pass
            continue
    
    if not success:
        # Last resort: try without any codec specification
        try:
            print("Trying final fallback: default settings...")
            output_video_path_fallback = output_video_path.replace('.mp4', '_fallback.avi')
            cropped_clip.write_videofile(
                output_video_path_fallback,
                verbose=False,
                logger=None
            )
            output_video_path = output_video_path_fallback
            success = True
            print(f"✓ Success with fallback format: {output_video_path}")
        except Exception as e:
            clip.close()
            cropped_clip.close()
            raise RuntimeError(f"All encoding attempts failed. Last error: {e}")
    
    # Clean up
    clip.close()
    cropped_clip.close()
    
    # Validate the output file
    if os.path.exists(output_video_path) and os.path.getsize(output_video_path) > 1000:  # At least 1KB
        print(f"✓ Successfully created cropped video: {output_video_path}")
        print(f"File size: {os.path.getsize(output_video_path) / (1024*1024):.1f} MB")
        
        # Try to validate by opening with MoviePy
        try:
            test_clip = VideoFileClip(output_video_path)
            print(f"✓ Output validation successful - Duration: {test_clip.duration:.2f}s")
            test_clip.close()
        except Exception as e:
            print(f"⚠ Warning: Output file created but validation failed: {e}")
        
        return output_video_path
    else:
        raise RuntimeError("Output file was not created or is too small")

# Updated preview function
def crop_video_with_preview_robust(input_video_path, preview_time=5, **crop_kwargs):
    """
    Show preview and crop with robust encoding.
    """
    from moviepy.editor import VideoFileClip
    import matplotlib.pyplot as plt
    import matplotlib.patches as patches
    
    # [Same preview code as before...]
    clip = VideoFileClip(input_video_path)
    preview_time = min(preview_time, clip.duration - 1)
    frame = clip.get_frame(preview_time)
    original_width, original_height = clip.size
    
    # Calculate conversions
    pixels_per_cm = crop_kwargs.get('pixels_per_cm', 50)
    left_cm = crop_kwargs.get('left_cm', 0)
    right_cm = crop_kwargs.get('right_cm', 0)
    top_cm = crop_kwargs.get('top_cm', 0)
    bottom_cm = crop_kwargs.get('bottom_cm', 0)
    
    left_px = int(left_cm * pixels_per_cm)
    right_px = int(right_cm * pixels_per_cm)
    top_px = int(top_cm * pixels_per_cm)
    bottom_px = int(bottom_cm * pixels_per_cm)
    
    # Show preview
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    ax1.imshow(frame)
    ax1.set_title(f'Original Frame (t={preview_time}s)')
    ax1.set_xlabel(f'Width: {original_width}px')
    ax1.set_ylabel(f'Height: {original_height}px')
    
    crop_rect = patches.Rectangle(
        (left_px, top_px), 
        original_width - left_px - right_px,
        original_height - top_px - bottom_px,
        linewidth=3, edgecolor='red', facecolor='none', linestyle='--'
    )
    ax1.add_patch(crop_rect)
    
    cropped_frame = frame[top_px:original_height-bottom_px, 
                         left_px:original_width-right_px]
    ax2.imshow(cropped_frame)
    ax2.set_title('Cropped Preview')
    ax2.set_xlabel(f'Width: {original_width-left_px-right_px}px')
    ax2.set_ylabel(f'Height: {original_height-top_px-bottom_px}px')
    
    plt.tight_layout()
    plt.show()
    
    clip.close()
    
    response = input("\nProceed with cropping? (y/n): ").lower().strip()
    if response in ['y', 'yes']:
        return crop_video_edges_robust(input_video_path, **crop_kwargs)
    else:
        print("Cropping cancelled.")
        return None
```
